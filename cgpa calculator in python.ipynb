```python
gradeValue = {"o":10,"a+":9,"a":8,"b+":7,"b":6,"ra":0}
finalCredits = 0
cgpa = 0
print("Calculate your CGPA")
name = input("please Enter your name: ")
noOfSem = int(input("Enter how many semester cgpa has to be found of: "))
print()
for i in range(1,noOfSem+1):
    gpa = 0
    totalCredits = 0
    print("Enter the details of the semester",i)
    noOfCourse = int(input("No of course you have registered: "))
    for j in range(noOfCourse):
        credit = int(input("Enter the credits for a subject: "))
        grade = input("Enter the grade: ").lower()
        totalCredits+=credit
        gpa+= credit*gradeValue[grade]
    print()    
    print("GPA for semester",i,":")     
    print("  --------------------------------------------------------------------------")    
    print("|  You have registered for total credits:",totalCredits,"and you have acquired GPA:",round(gpa/totalCredits,2)," |")
    print("  --------------------------------------------------------------------------")
    print()
    finalCredits+=totalCredits
    cgpa += gpa/totalCredits    
print(name,"your CGPA is:",round(cgpa/noOfSem,2))    
print()    
print("  ---------------------------------------------------------------------------")    
print("|  You have registered for total credits:",finalCredits,"and you have acquired CGPA:",round(cgpa/noOfSem,2)," |")   
print("  ---------------------------------------------------------------------------")  
```
